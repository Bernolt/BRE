```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/alcohol.csv')
```

```python
df = df.rename(columns={'Unnamed: 0': 'name'}).set_index('name')
```

```python
df = df.fillna(0)
```

```python
df
```

```python
df.index
```

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
np.round(euclidean_distances(df, df), 2)
```

```python
sim = pd.DataFrame(
    np.round(euclidean_distances(df, df), 2),
    columns=df.index, 
    index=df.index)

sim
```

```python
sim.loc['Max'].sort_values(ascending=True)[1:]
```
